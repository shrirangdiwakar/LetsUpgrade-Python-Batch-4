## **Silver Project 1 : Tic-Tac-Toe**

```
#Entire Code is written by ME(Shrirang Diwakar) , no reference from anywhere ! Code is less than 100 Lines in actual case!

import random

choice ='yes'


#function for displaying board
def disp_board(seq):
    print("\n\t   ",seq[0],"\t|  ",seq[1],"\t|   ",seq[2],"\n\t--------------------------\n\t   ",seq[3],"\t|  ",seq[4],"\t|   ",seq[5],"\n\t--------------------------\n\t   ",seq[6],"\t|  ",seq[7],"\t|   ",seq[8])



#function for mapping numpad
def map_num(x):
  if(x==1):
    return 6
  elif(x==2):
    return 7
  elif(x==3):
    return 8
  elif(x==4):
    return 3
  elif(x==5):
    return 4
  elif(x==6):
    return 5
  elif(x==7):
    return 0
  elif(x==8):
    return 1
  else :
    return 2




#function for WinLose Check : all the possible winning conditions are displayed
def win_lose(seq,x,i):

    if   (seq[0]==seq[1] and seq[1]==seq[2]) and (seq[0]!='' and seq[1]!='' and seq[2]!=''):
      print("\033[32m\033[1mHurray, Player ",i," Won the game\033[0m")
      return i
    elif (seq[3]==seq[4] and seq[4]==seq[5]) and (seq[3]!='' and seq[4]!='' and seq[5]!=''):
      print("\033[32m\033[1mHurray, Player ",i," Won the game\033[0m")
      return i
    elif (seq[6]==seq[7] and seq[7]==seq[8]) and (seq[6]!='' and seq[7]!='' and seq[8]!=''):
      print("\033[32m\033[1mHurray, Player ",i," Won the game\033[0m")
      return i
    elif (seq[0]==seq[3] and seq[3]==seq[6]) and (seq[0]!='' and seq[3]!='' and seq[6]!=''):
      print("\033[32m\033[1mHurray, Player ",i," Won the game\033[0m")
      return i
    elif (seq[1]==seq[4] and seq[4]==seq[7]) and (seq[1]!='' and seq[4]!='' and seq[7]!=''):
      print("\033[32m\033[1mHurray, Player ",i," Won the game\033[0m")
      return i
    elif (seq[2]==seq[5] and seq[5]==seq[8]) and (seq[2]!='' and seq[5]!='' and seq[8]!=''):
      print("\033[32m\033[1mHurray, Player ",i," Won the game\033[0m")
      return i
    elif (seq[0]==seq[4] and seq[4]==seq[8]) and (seq[0]!='' and seq[4]!='' and seq[8]!=''):
      print("\033[32m\033[1mHurray, Player ",i," Won the game\033[0m")
      return i
    elif (seq[2]==seq[4] and seq[4]==seq[6]) and (seq[2]!='' and seq[4]!='' and seq[6]!=''):
      print("\033[32m\033[1mHurray, Player ",i," Won the game\033[0m")
      return i
    else:
      return 0





#main function working
while choice!='no':
  print("\n\n\t\033[1m\033[90m*****WELCOME TO TIC-TAC-TOE by Shrirang Diwakar*****\033[0m\n\n")

#name input
  p1 = input("Enter name of Player 1 :")
  p2 = input("Enter name of Player 2 :")


#declaration
  seq = ['','','','','','','','','']
  free_space = 9
  flag=0

#random
  f = random.randint(0,1)
  if f==0 :
    p11 = p1
    p22 = p2
  else :
    p11 = p2
    p22 = p1

  print("\033[32m\nPLAYER",f+1," i.e ",p11.upper()," WILL GO FIRST")


  print("\n\n\033[94m***\033[4mSymbol\033[0m\033[94m***\033[0m")
  print("",p11.upper(),"(PLAYER ",f+1,"): \033[1mX\n\033[0m",p22.upper(),"(Player ",int(not(f))+1,"): \033[1mO")


  print("\nNote: For Tic-Tac-Toe use Numpad as the positions for the Board and Enter the numbers as input e.g.")
  print("\n\t    7\t|   8\t|    9\n\t--------------------------\n\t    4\t|   5\t|    6\n\t--------------------------\n\t    1\t|   2\t|    3\033[0m")
  
  
#logic for alternate turns,mapping numpad,win/lose  
  while free_space!=0 :

    
    if free_space % 2 == 1 :                                                    #to check player 1 or player 2
      x = int(input("\n {}'s Turn ,Make your Move :\t".format(p11)))
      x = map_num(x)                                                            #mapping numpads
 
      if seq[x]=='':
        seq[x] = 'X'
        free_space-=1
        disp_board(seq)
        i = f+1
        if free_space<7 :                                                       #min 3 moves are required for winning
          if win_lose(seq,x,i):
            flag=1
            print("\033[1m\033[32mCongrats ",p11)
            print("\033[1m\033[91mHard Luck ",p22)
            break
          
      else :
        print("\n\n\033[91m\033[1mAlready Filled position, Try Again\033[0m")
  
    else :
      x = int(input("\n {}'s Turn ,Make your Move :\t".format(p22)))
      x = map_num(x)
    
      if seq[x]=='':
        seq[x] = 'O'
        free_space-=1
        disp_board(seq)
        i = int(not(f))+1
        if free_space<7 :
          if win_lose(seq,x,i):
            flag=1
            print("\033[1m\033[32mCongrats ",p22)
            print("\033[1m\033[91mHard Luck ",p11)
            break
      else :
        print("\n\n\033[91m\033[1mAlready Filled position, Try Again\033[0m")

  if flag==0:
    print("\033[1m\033[94mOops, Its a TIE\033[0m")
  choice = input("\n\nWANT TO PLAY AGAIN ? ENTER 'YES' or 'NO':\t")
  choice.lower()

```
